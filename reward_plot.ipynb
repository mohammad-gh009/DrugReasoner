```python
import json
import matplotlib.pyplot as plt


with open('trainer_state_300.json') as f:
    data = json.load(f)
```

```python
# Extract steps and rewards
steps = [entry['step'] for entry in data['log_history']]
rewards = [entry['rewards/strict_format_reward_func'] for entry in data['log_history']]

# Plotting
plt.figure(figsize=(8,5))
plt.plot(steps, rewards, marker='o', linewidth=1, color='#14c9e2')
plt.xlabel('Step', fontsize=16)
plt.ylabel('Reward', fontsize=16)
plt.grid(True, alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
steps = [entry['step'] for entry in data['log_history']]
rewards = [entry['reward'] for entry in data['log_history']]

# Normalize the rewards to the range of -1 to 1
min_reward = min(rewards)
max_reward = max(rewards)
normalized_rewards = [(2 * (reward - min_reward) / (max_reward - min_reward)) - 1 for reward in rewards]

# Plotting
plt.figure(figsize=(8,5))
plt.plot(steps, normalized_rewards, marker='o', linewidth=1, color='#14c9e2')
plt.xlabel('Step', fontsize=16)
plt.ylabel('Normalized Reward', fontsize=16)
plt.ylim(-1, 1)  # Set y-axis limits
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```
