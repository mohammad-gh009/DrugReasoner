```python
import json
import matplotlib.pyplot as plt


with open('trainer_state_200.json') as f:
    data = json.load(f)
```

```python
# Extract steps and rewards
steps = [entry['step'] for entry in data['log_history']]
rewards = [entry['reward'] for entry in data['log_history']]

# Plotting
plt.figure(figsize=(8,5))
plt.plot(steps, rewards, marker='o', linewidth=1, color='#14c9e2')
plt.xlabel('Step', fontsize=16)
plt.ylabel('Reward', fontsize=16)
plt.grid(True, alpha=0.1)
plt.tight_layout()
plt.show()
```
